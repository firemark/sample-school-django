```python
schools.groupby('city')[
    ['boys', 'girls', 'fulltime_teachers']
].agg(['mean', 'sum']).sort_values(('boys', 'sum'), ascending=False)
```

```python
schools['students'] = schools.apply(lambda r: r['boys'] + r['girls'], axis=1)
```

```python
import matplotlib
%matplotlib inline
(
    schools[schools.school_code == 'Gimnazjum']
    .groupby('city')
    .sum()[['students', 'girls', 'boys']]
    .sort_values('students', ascending=False)[:20]
    .plot(kind='bar', figsize=(15, 5))
)
```

```python
(
    schools.groupby('school_code')
    .sum()[['students', 'girls', 'boys']]
    .sort_values('students', ascending=False)[:5]
    .plot(kind='bar', figsize=(15, 5))
)
```
